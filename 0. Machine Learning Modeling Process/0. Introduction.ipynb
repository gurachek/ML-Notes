# Building a Machine Learning Portfolio

https://machinelearningmastery.com/build-a-machine-learning-portfolio/

- Each project has a clearly defined start and end point.
- You will learn something from the project
- Go for skills / try new things as often as possible
- Practice your communication and delivery of results

## Properties of a Good Portfolio

- Accessible (GitHub)
- Small Tasks (10-20 Hours each)
- Modest Goals (allows for project completion)
- Understandable


## Project Ideas

- Investigate a property of a machine Learning Tool
- Investigate behavior of an algorithm
- Investigate a dataset
- Implement an algorithm
