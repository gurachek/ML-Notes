## 2. Learning = Representation + Evaluation + Optimization
When you have a problem that could be solved with machine learning the trick to navigating the wasteland of hundreds of potential machL algorithms is to remember that this space is a combination of three components.

1. Representation- how is the classifier represented in a formal language that a computer can handle? Choosing a representation determines what potential classifiers a learner can become. This set of potential classifiers is the hypothesis space of the learner.

2. Evaluation - Scoring/ objective function distinguishes good classifiers from bad ones. Does the algorithm have an internal objective function that matches the external goal which we wish to optimize?

3. Optimization - Need a method to search among classifiers for the highest scoring one.Optimization technique sets the efficiency of the learner


## 4. Data Alone is Not Enough

If you are trying to model a lot of variables, the data needed grows exponentially. For learning a boolean function with 100 variables, you need $2^{100}$ examples. Every learner must have asumptions beyond the data its given in orger to generalize. 

Basically you are trying to model an unknnown function without perfect information. At some point guessing is more accurate then anything a learner could become.

The functions we want to learn can have very basinc assumptions (smoothness, similar examples have limilar classes, limited dependencies, or limited complexity.) These simple assumptions allow machine learning to do WELL ENOUGH to be useful in certain situations. If you pick your battles and make the right assumptions you dont need god like amountss of data.

Learners learn through induction, input examples create a large amount of output knowledge. (Think Marco Polo)  The more data we put in, the more we will get out.

Choosing a representation for a learner is largely based off of what kind of knowledge are easily expressed by it.

## 5. Overfitting Has Many Faces.

One Way of decomposing overfitting is bias vs variance. Bias is learners tendency to consistently learn the same wrong thing. Variance is learner learning random things regardless of the real signal. Similar to fishing for one fish with a net vs a harpoon. variance is issues with casting too wide a net, and bias is the issue of missing the target with your harpoon.

Very rigid models have higher bias since they cant get on target with the true function. Flexible models have a high variance because they become too flexible and cast a wider prediction aas a result of flexing. In terms of optimization

A more powerful (flexible?) learber i not necessarily better.  Strong false assumptions at times can perform better than weak true assumptions. mainly because more flexible learners require more data to avoid over fitting. The correct choice of algorithm depends on the amount of data and the assumptions you should make on what we can do to reduce variance and bias. Rule of thumb is produce more accurate results by lowering bias with flexible models and reduce variance by shoveling in a crap ton of data. Given a fixed amount of data, you must choose what amount of flexibility keeps bias low without overfitting.

#### Ways to Combat Overfitting

Cross fitting can help with hyperparameter tuning, but too much of this leads to overfitting as well. 

Another way is adding a regularization term to your evaluation function. this might penalize potential classifiers with more structure by adding certain constraints. Puts a limit on how much we can reduce variance and flex to data. 

Avoiding overfitting (variance) is a balancing act, dont go to far or else you risk underfitting (bias)


#### Back to Overfitting

Overfitting is not just caused by the presence of noise in our data. Overfitting can be similar to the problem of multiple testing in statistics. If we have enough trial runs of potential classifiers, one might get lucky but not actually be a good classifier on the real data. We have to make sure the performance of the classifier is significant and not just luck. Adjust the significance threshold based on the number of hypothesis being tested, tantamount to reducing the false discovery rate.













## 6. Intuition Fails in High Dimensions

Many algorithms that work in low dimensions are impossible in high dimensions. Generalizzing becomes exponentially harder as the number of features grows. Because a fixed number of training samples occupies a smaller and smaller fraction of the input space. Essentially finding a certain needle in a hay stack vs finding a certain needle in New York. Each training example gives us a glimpse of our surroundings, but finding something in New York requires many many observations.

The similarity assumption breaks down in high dimension. A nearest neighbor in high d might have certain features that signal a true output, unfortunately alll of the other features might produce enough noise output that a correct classification might be drowned out.

As the dimensionalty increases so do the number of nearest neighbors since space expands. think 3 feet vs 3 feet$^{2}$ one has a significantly higher spatial containment. THus in high enough D every example might seem similar, so how do we classify?

## L1 vs L2 Regularization
The regularized term lies on one of the two curves.



### L1 Regularization
L1 regularization is based off of absolute difference. The regularization is in linear proportion to the absolute distance.

"Contrarily, L1 regularization forces non-essential entries of theta all the way to zero."


### L2 Regularization
Based off of squared distance. Pulled towards zero in proportion to the square distance. 

"The result is that L2 regularization drives many of your parameters down, but will not necessarily eradicate them, since the penalty all but disappears near zero."

### Ridge (shrink the coeffs)
Uses L2 regularization - vulnerable to outliers, limits the size of the coefficient vector, but does not creat a sparse coefficient vector. computationally less expensive. 

### LASSO  (sparse coeffs)

uses L1 regularization more robust, butless computationally efficient. Designed to create sparse vector ~> feature engineering is built in. one does not know which variables are put in.

Vulnerable to correlated data!!! if this is present, use elastic net.

### Elastic Net (shrink and sparse)

Compromise between the two uses both L1 and L2. Use when collinearity is expected.
