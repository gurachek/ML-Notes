### Precision vs Recall

Precision - of the times we said yes, what percentage of the time were we right?
Recall - of the times we should have said yes, what percentage of the time did we?

High precision -> When we say yes we are right.
High Recall -> when we need to say yes, we say yes.

obviously one concept funnels into the other, so a good model has both precision and recall.

The F1 score is the harmonic mean of precision and recall

F2 measures recall higher than precision by placing more emphasis on false negatives

Both F-Score are common in the NLP Literature
